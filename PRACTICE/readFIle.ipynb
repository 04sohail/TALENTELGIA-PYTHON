```python
# =>You have to create a method which will update the excel file with the given paramters.
# 1) column name => which column data to update.
# 2) row vaue => which row value to update.

def changeValue (rowNumber, columnName,  valueToUpdate):
    if rowNumber in df.index and columnName.lower() in pd.Series(df.columns).str.lower().values:
        print("Cell Found")
        df.loc[rowNumber, columnName]= valueToUpdate
        print("Value Changed Successfully")
        print("-"*50)
    else:
        print("Invalid Row Number Or Column Name !!!")
        return
rowNumber = int(input("Enter Row Number : "))
columnName = input("Enter Column Name : ").strip()
valueToUpdate = input("Enter Value To Update : ").strip()
changeValue(rowNumber, columnName, valueToUpdate)
```

```python
# => Also create a method which will find the avg sales by country in a year. Using these paramters.
# 1) country name
# 2) year name
# 3) product name optional.
# =>If user is providing product name than find the avg sales of that product in that year.
```

```python
def findAverageSales(countryName, year, productName = ""):
    if countryName in df["Country"].str.lower().values and year in df["Year"].values:
        if productName == "":
            average = df[(df["Year"] == year) & (df["Country"].str.lower().values == countryName)][" Sales"].mean()
            print("The Average Sales Of {} Country In Year {} = {}.".format(countryName, year, average))
        else:
            if productName in df["Product"].str.lower().values:
                average = df[(df["Year"] == year) & (df["Country"].str.lower().values == countryName) & (df["Product"].str.lower().values == productName)][" Sales"].mean()
                print("The Average Sales Of {} Country In Year {} Of Product {} = {}.".format(countryName, year, productName,average))
            else:
                print("Enter Valid Product Name !!!")
                return
    else:
        print("Enter Valid Country Name Or Year !!!")
        return
try:        
    countryName = input("Enter Country Name : ").strip().lower()
    year = int(input("Enter Year  : "))
    productName = input("Enter Prouct Name (Optional) : ").strip().lower()
    findAverageSales(countryName, year, productName)
except ValueError:
    print("Enter Valid Year !!!")
```

```python
# =>Similarly create the method to find total sales by country in a year. Product is also optional here
def totalSales(countryName, year, productName = ""):
    if countryName in df["Country"].str.lower().values and year in df["Year"].values:
        if productName == "":
            total = df[(df["Year"].values == year) & (df["Country"].str.lower().values == countryName)][" Sales"].sum()
            print("Total Sales Of {} Country In Year {} = {}.".format(countryName, year, total))
        else:
            if productName in df['Product'].str.lower().values:
                total = df[(df["Year"].values == year) & (df["Country"].str.lower().values == countryName) & (df["Product"].str.lower().values == productName)][" Sales"].sum()
                print("The Average Sales Of {} Country In Year {} Of Product {} = {}.".format(countryName, year, productName,total))
            else:
                print("Enter Valid Product Name !!!")
                return
    else:
        print("Enter Valid Country Name Or Year !!!")
        return
try:        
    countryName = input("Enter Country Name : ").strip().lower()
    year = int(input("Enter Year  : "))
    productName = input("Enter Prouct Name (Optional) : ").strip().lower()
    totalSales(countryName, year, productName)
except ValueError:
    print("Enter Valid Year !!!")  
```
